```
# @title Import all modules
from transformers import pipeline
import gradio as gr
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
classifier = pipeline("text-classification",model="distilbert/distilbert-base-uncased-finetuned-sst-2-english")
```

```
classifier("I think the product is not all that great")
```

```
import gradio as gr

def predict_sentiment (text):
  return f"The sentiment seems to be {classifier(text)[0]['label']} with a score of {classifier(text)[0]['score']}."

gr.Interface(fn=predict_sentiment,inputs="text",outputs="text").launch()
```
